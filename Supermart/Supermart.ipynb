```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("Supermart Grocery Sales .csv")
print(data)
```

```python
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
```

```python
data.head()
```

```python
# Profit by Category
category_profit = data.groupby('Category').agg(
    Total_Profit=('Profit', 'sum'),
    Average_Profit=('Profit', 'mean')
).reset_index()

# Profit by Sub-Category
sub_category_profit = data.groupby('Sub Category').agg(
    Total_Profit=('Profit', 'sum'),
    Average_Profit=('Profit', 'mean')
).reset_index()

# Profit by Region
region_profit = data.groupby('Region').agg(
    Total_Profit=('Profit', 'sum'),
    Average_Profit=('Profit', 'mean')
).reset_index()

# Profit by Cities
city_profit = data.groupby('City').agg(
    Total_Profit=('Profit', 'sum'),
    Average_Profit=('Profit', 'mean')
).reset_index()

# Sales by Region
region_sales = data.groupby('Region').agg(
    Total_Sales=('Sales', 'sum'),
    Average_Sales=('Sales', 'mean')
).reset_index()

# Sales by Cities
city_sales = data.groupby('City').agg(
    Total_Sales=('Sales', 'sum'),
    Average_Sales=('Sales', 'mean')
).reset_index()

print('CATEGORY DETAILS')
print(category_profit.head())
print("------------------------------------------------------")
print('SUB CATEGORY DETAILS')
print(sub_category_profit.head())
print("------------------------------------------------------")
print('REGION-WISE PROFIT')
print(region_profit)
print("------------------------------------------------------")
print('CITY-WISE PROFIT')
print(city_profit)
print("------------------------------------------------------")
print('REGION-WISE SALES')
print(region_sales)
print("------------------------------------------------------")
print('CATEGOREGION-WISE SALES')
print(city_sales)
print("------------------------------------------------------")
```

```python
sns.set(style = 'darkgrid')
```

```python
plt.figure(figsize=(15,6))
sns.histplot(data['Sales'], bins = 30, kde = True, color = 'Green')
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
data.head()
```

```python
# Extract the month
data['Month'] = data['Order Date'].dt.to_period('M')

# Group by Month and aggregate Sales
monthly_sales = data.groupby('Month')['Sales'].sum().reset_index()
# Convert 'Month' from Period to string for plotting
monthly_sales['Month'] = monthly_sales['Month'].astype(str)

# Group by Month and aggregate Profit
monthly_profit = data.groupby('Month')['Profit'].sum().reset_index()
# Convert 'Month' from Period to string for plotting
monthly_profit['Month'] = monthly_profit['Month'].astype(str)
```

```python
plt.figure(figsize=(15, 6))
sns.lineplot(x='Month', y='Sales', data=monthly_sales, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(15, 6))
sns.lineplot(x='Month', y='Profit', data=monthly_profit, marker='o', color = "Red")
plt.title('Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.xticks(rotation=90)
plt.show()
```
